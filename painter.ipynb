```python
import matplotlib.pyplot as plt  
import os
import json

os.system("python3 generator.py 47 poisson")

plt.figure(figsize = (20, 10))

policies = ['min-add-time', 'best-fit']
for policy in policies:
    os.system("python3 simulator.py {}".format(policy))
    data = json.load(open("output.json"))
    X = []
    Y = []
    for elem in data:
        if elem['time'] >= 20000 and elem['time'] <= 100000:
            X.append(elem['time'])
            Y.append(elem['active hosts'])
    plt.title("Active hosts over time", fontsize = 24)
    plt.plot(X, Y, label = policy)
    plt.legend(fontsize = 20)
```
