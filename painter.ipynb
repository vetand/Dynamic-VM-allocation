```python
import matplotlib.pyplot as plt  
import os
import json

os.system("python3 generator.py 47 poisson")

plt.figure(figsize = (20, 10))

plt.vlines(x = 15000, ymin = 0, ymax = 80.0, colors= 'purple', ls = '--', lw = 2, 
                                                        label = 'simulation begins')
plt.vlines(x = 75000, ymin = 0, ymax = 80.0, colors= 'purple', ls = '--', lw = 2, 
                                                        label = 'simulation ends')

policies = [['min-add-time-2', 'non-adaptive'], ['best-fit', 'non-adaptive']]
for policy in policies:
    os.system("python3 simulator.py {} {}".format(policy[0], policy[1]))
    data = json.load(open("output.json"))
    X = []
    Y = []
    for elem in data:
        X.append(elem['time'])
        Y.append(elem['active hosts'])
    plt.title("Active hosts over time", fontsize = 24)
    plt.plot(X, Y, label = policy)

# ideal result in terms of number of bins used, simply the ideal_bins_number = int(total_util) + 1
    
#os.system("python3 simulator.py {}".format(policy, 'non-adaptive'))
#data = json.load(open("output.json"))
#X = []
#Y = []
#for elem in data:
#    X.append(elem['time'])
#    Y.append(elem['ideal result'])
#
#plt.plot(X, Y, label = 'ideal result')
plt.legend(fontsize = 20)
```
